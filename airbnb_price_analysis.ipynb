# Airbnb Price Analysis

```python
# Loading the dataset
import pandas as pd

file_path = "Data/Airbnb_Data.csv"
data = pd.read_csv(file_path)

# Display the first 5 rows
print(data.head())
```

```python
# Checking the structure of the dataset
print(data.info())
print(data.describe())
```

```python
# Data cleaning
print(data.isnull().sum())
```

# Exploratory Data Analysis

```python
# Visualizing the data to find patterns or anomalies using matplotlib or seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Filtering out non-numeric columns
numeric_data= data.select_dtypes(include=['number'])
# Computing the correlation matrix for numeric columns
corr_matrix= numeric_data.corr()

# Example: Distribution of a numeric column
plt.figure(figsize=(12,6))
sns.histplot(data['log_price'], bins=30, kde=True)
plt.savefig("histoplot.png")
plt.show()
plt.close()
```

```python
# Looking for correlations

plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.savefig("heatmap.png")
plt.show()
plt.close()
```
